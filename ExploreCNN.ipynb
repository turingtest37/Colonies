```julia
using Images
using Flux
using FileIO
using Statistics
```

```julia
function normalize_colony(file::AbstractString)
    img = FileIO.load(file)

    imgg = Gray.(img)

    sigma = std(real(imgg))

    f = Kernel.gaussian(6 * sigma)

    imgg = imfilter(f, imgg)

    img = imresize(imgg,100,100)

    return img
end
```

```julia
img = FileIO.load("./img/best/colony4j-285ddb84-0869-11e8-3375-c3e7001ae19f.png")
```

```julia
imgg = Gray.(img)
```

```julia
sigma = 2.5
```

```julia
f = Kernel.gaussian(sigma)
```
